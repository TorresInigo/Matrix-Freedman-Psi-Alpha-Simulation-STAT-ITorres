# <u> **Bandit Simulation**</u>

```python
# ============================================================
# Colab-ready: MAX-EVIDENCE NOTEBOOK CELL
#  - Crossover helpers (Freedman–ψ_α vs K&S(Ber))
#  - Head-to-head bandit sims (Freedman vs K&S), paired
#  - Uses your paper’s constants exactly (√(32/3), ξ(x) with \tilde C_α, etc.)
# ============================================================
# [Paper linkage]
# • Section "Algorithmic and Statistical Applications" → Bandit–Learning Consequences (§ bandit)
# • Eq. (self-norm-weibull)–(beta-oful-weibull): β^{(Weib)} = c1 * sqrt(2 * log(det term / δ))
#   + c2 * ν * (log(c3 n / δ))^{1/α}, with c1 = √(32/3), c2 = 2·4^{1/α}, c3 = 6·\tilde C_α
# • Proposition in §KS-comparison: θ_F(x) vs θ_{KS}^{(Ber)}(x) and the crossover conditions
# • Notation: z = (4 log(eU/σ))^{1/α} and A = (3/4) K z for the K&S(Ber) envelope
# • θ_F(x) combines the Freedman (quadratic) piece and the "big-jump/drift" tail piece via ξ(x)

import math, numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass
from typing import Optional

# =========================
# =======  CONFIG  ========
# =========================
# [Paper linkage] Simulation controls for the bandit experiments summarized in Tables (α=0.9, 1.3, 2.0),
# dimension d, horizon n, candidate set size per round, seeds, etc. (cf. §Simulation Study: Linear bandits).

DELTA: Optional[float] = None     # if None, fall back to 1e-6 for the bandit run
# (σ, U, K) are *only* for the K&S(Ber) baseline
SIGMA_KS = 0.0
U_KS     = 0.0
K_KS     = 0.0

# Bandit simulation
RUN_BANDITS          = True
T_ROUNDS             = 100_000     # rounds per seed
D_BANDIT             = 20
S_NORM               = 1.0         # ||θ*|| bound
L_BOUND              = 1.0         # ||x_t|| bound
LAMBDA               = 1.0         # ridge
N_SEEDS              = 10          # increase on a cluster if you like
CANDS_PER_ROUND      = 512         # candidate actions per round
STORE_FULL_TRAJECTORY = True       # if False, only store end-regret

# Keep K&S bandit params in sync with current test set
BANDIT_SIGMA = SIGMA_KS
BANDIT_U     = U_KS
BANDIT_K     = K_KS

# ================================
# == Theory helpers from paper ==
# ================================
# [Paper linkage] The next helpers are direct programmatic counterparts of your mathematical
#   constants and envelopes in the paper:
# - \tilde C_α (Remark/Theorem constants in main result; drift coefficient)
# - ξ(x)   (Prop. in §KS-comparison; collects union-bound logs for big-jump & drift control)
# - θ_F(x) (Freedman–ψ_α threshold: quadratic 'Freedman piece' + heavy-tail 'tail piece')
# - z, A, θ_KS(x)  (K&S(Ber) structure; Theorem 2.1 style envelope with linear A·x term)
# - ρ, ρ_min, x_* and x_lb_logs (compatibility and algebraic thresholding used in the prop.)

def tilde_C_alpha(alpha: float) -> float:
    """
    Your tight drift constant  \tilde C_α = 2(2/α)^{1/α} e^{-1/α} + 2^{1+1/α} Γ(1/α)/α
    (Theorem statement + proof remarks in your revision.)
    """
    if alpha <= 0.0 or alpha > 2.0:
        raise ValueError("alpha must be in (0, 2].")
    return 2.0*(2.0/alpha)**(1.0/alpha) * math.e**(-1.0/alpha) + (2.0**(1.0+1.0/alpha))*math.gamma(1.0/alpha)/alpha

def xi_of_x(x: float, n: int, C_tilde: float) -> float:
    """
    Your ξ(x) from Prop. (KS-comparison), written with \tilde C_α:
        ξ(x) = max{ x + log(6n),  2 ( x + log(6 \tilde C_α n) + 1 ) }.
    """
    return max(x + math.log(6.0*n), 2.0*(x + math.log(6.0*C_tilde*n) + 1.0))

def theta_F(alpha: float, nu: float, v: float, n: int, x: float, C_tilde: float) -> float:
    """
    θ_F(x) = sqrt{ (32/3) v (x + log 3) } + max{ 2 ν [ (v ξ(x))/ν^2 ]^{1/α},  e ν }.
    First term = Freedman piece (3/32 exponent ↔ √(32/3) in radii);
    second term enforces the big-jump/drift constraints with your \tilde C_α in ξ(·).
    """
    part_quad = math.sqrt((32.0/3.0)*v*(x + math.log(3.0)))
    xi = xi_of_x(x, n, C_tilde)
    part_tail = max(2.0*nu*((v*xi)/(nu**2))**(1.0/alpha), math.e*nu)
    return part_quad + part_tail

def z_KS(alpha: float, U: float, sigma: float) -> float:
    # K&S(Ber) factor  z = (4 log(e U / σ))^{1/α}.
    return (4.0 * max(1e-12, math.log(math.e * U / max(sigma, 1e-12)))) ** (1.0/alpha)

def A_KS(alpha: float, K: float, U: float, sigma: float) -> float:
    # K&S(Ber) linear coefficient:  A = (3/4) K z
    return 0.75 * K * z_KS(alpha, U, sigma)

def theta_KS(alpha: float, sigma: float, K: float, U: float, x: float) -> float:
    # θ_KS(x) = σ√(2x) + A x
    A = A_KS(alpha, K, U, sigma)
    return sigma*math.sqrt(2.0*x) + A*x

def rho(v: float, nu: float, alpha: float) -> float:
    # Compatibility statistic ρ = (v/ν^2)^{1/α}
    return (v/(nu**2)) ** (1.0/alpha)

def rho_min(x_: float) -> float:
    # Need ρ ≥ (32/3)(x + log 3) so Freedman’s exponent ≥ x + log 3
    return (32.0/3.0)*(x_ + math.log(3.0))

def x_lb_logs(n: int, C_tilde: float) -> float:
    # Lower bound for x collecting the union-bound logs from big-jump & drift
    return max(math.log(6.0*n), 2.0*(math.log(6.0*C_tilde*n)+1.0), 1.0, math.log(3.0))

def x_star(alpha: float, nu: float, v: float, n: int, sigma: float, U: float, K: float) -> float:
    """
    For α>1: x_* is the max of the algebraic constraints from your Prop. (KS-comparison),
    ensuring θ_F(x) ≤ θ_KS(x) at confidence e^{-x} (given the compatibility on ρ).
    """
    C_tilde = tilde_C_alpha(alpha)
    A = A_KS(alpha, K, U, sigma)
    base = x_lb_logs(n, C_tilde)
    term1 = (256.0/3.0) * (v/(A**2))
    term2 = ((4.0*(4.0**(1.0/alpha)) * (nu**(1.0-2.0/alpha)) * (v**(1.0/alpha)))/A) ** (alpha/(alpha-1.0))
    term3 = (2.0*math.e*nu)/A
    return max(base, term1, term2, term3)

def crossover_report(alpha: float, nu: float, v: float, n: int, d: int,
                     sigma: float, U: float, K: float,
                     x_direct: Optional[float]=None, delta: Optional[float]=None):
    """
    Returns a dict summarizing θ_F vs θ_KS at a chosen x.
    - α>1: uses x_* from your Prop. to guarantee θ_F ≤ θ_KS under the ρ-compatibility.
    - α≤1: returns an admissible [x_lo, x_hi]; midpoint is a convenient choice.
    """
    C_tilde = tilde_C_alpha(alpha)
    x_from_delta = (math.log(d/float(delta)) if (delta is not None) else None)
    x_chosen = x_direct if x_direct is not None else x_from_delta

    if alpha > 1.0:
        xs = x_star(alpha, nu, v, n, sigma, U, K)
        if x_chosen is None or x_chosen < xs:
            x_chosen = xs
        compat = rho(v, nu, alpha) >= rho_min(x_chosen)
        thF = theta_F(alpha, nu, v, n, x_chosen, C_tilde)
        thK = theta_KS(alpha, sigma, K, U, x_chosen)
        return dict(alpha=alpha, nu=nu, v=v, n=n, d=d, sigma=sigma, U=U, K=K,
                    C_tilde=C_tilde, x=x_chosen, x_star=xs, x_lo=None, x_hi=None,
                    alpha_case=">1", rho=rho(v,nu,alpha), rho_min=rho_min(x_chosen),
                    compatibility_ok=compat, theta_F=thF, theta_KS=thK,
                    freedman_leq_KS=(thF<=thK and compat))

    # α ≤ 1 case
    A = A_KS(alpha, K, U, sigma)
    base = x_lb_logs(n, C_tilde)
    x_lo = base if x_from_delta is None else max(base, x_from_delta)
    denom = 2.0*(4.0**(1.0/alpha))*(nu**(1.0-2.0/alpha))*(v**(1.0/alpha))
    x_hi = (A/denom)**(alpha/(1.0-alpha)) if denom>0 else float("inf")
    if x_direct is not None:
        x_chosen = x_direct
    else:
        x_chosen = 0.5*(x_lo+x_hi) if x_lo < x_hi else x_lo
    compat = rho(v, nu, alpha) >= rho_min(x_chosen)
    thF = theta_F(alpha, nu, v, n, x_chosen, C_tilde)
    thK = theta_KS(alpha, sigma, K, U, x_chosen)
    return dict(alpha=alpha, nu=nu, v=v, n=n, d=d, sigma=sigma, U=U, K=K,
                C_tilde=C_tilde, x=x_chosen, x_star=None, x_lo=x_lo, x_hi=x_hi,
                alpha_case="<=1", rho=rho(v,nu,alpha), rho_min=rho_min(x_chosen),
                compatibility_ok=compat, theta_F=thF, theta_KS=thK,
                freedman_leq_KS=(thF<=thK and compat))

# =========================================
# === Bandit: generators & core routines ===
# =========================================
# [Paper linkage] Data generation follows the ψ_α noise model (symmetric sign·Exp(1)^{1/α} scaling),
# consistent with the i.i.d. ψ_α synthetic model used in your §Simulation Study (bandits).

def subweibull_sym(alpha: float, nu_lux: float, size: int, rng: np.random.Generator) -> np.ndarray:
    """
    Symmetric ψ_α noise with Luxemburg radius ≈ ν:
        E[exp(|Y|^α / ν^α)] ≈ 2.
    We use sign * Exp(1)^{1/α} scaling; see standard ψ_α constructions.
    """
    if alpha <= 0.0 or alpha > 2.0:
        raise ValueError("alpha must be in (0, 2].")
    nu_weib = nu_lux / (2.0 ** (1.0/alpha))
    W = rng.exponential(scale=1.0, size=size)
    mags = nu_weib * np.power(W, 1.0/alpha)
    signs = rng.choice([-1.0, 1.0], size=size)
    return signs * mags

def candidates_for_t(t: int, d: int, L: float, cands: int, seed_base: int) -> np.ndarray:
    # Same candidate set for both policies at round t (paired comparison)
    rng = np.random.default_rng(seed_base + t)
    X = rng.normal(size=(cands, d))
    X /= np.linalg.norm(X, axis=1, keepdims=True) + 1e-18
    X *= L
    return X

def noise_for_t(t: int, alpha: float, nu_lux: float, seed_base: int) -> float:
    # Same noise sample for both policies at round t (paired comparison)
    rng = np.random.default_rng(seed_base + t)
    return float(subweibull_sym(alpha, nu_lux, 1, rng)[0])

def _V_inv_times_X(V: np.ndarray, X: np.ndarray) -> np.ndarray:
    """
    Solve V Z = X for Z, batched (more stable than explicitly inverting V).
    Returns Z = V^{-1} X.
    """
    return np.linalg.solve(V, X.T).T

def beta_weib_freedman(alpha: float, nu_lux: float, V: np.ndarray, lam: float,
                       delta: float, n: int) -> float:
    """
    Self-normalized ψ_α bandit radius from your §Applications:
      β = c1 * sqrt( 2 * log( det(V)^{1/2} det(λI)^{-1/2} / δ ) )
          + c2 * ν * (log(c3 n / δ))^{1/α},
    with c1 = √(32/3), c2 = 2*4^{1/α}, c3 = 6*\tilde C_α.
    """
    sign, logdetV = np.linalg.slogdet(V)
    if sign <= 0:
        raise RuntimeError("Design matrix V must be SPD.")
    d = V.shape[0]
    logdetLamI = d*math.log(lam)
    xdet = 0.5*(logdetV - logdetLamI) - math.log(delta)
    xdet = max(xdet, 1e-12)
    c1 = math.sqrt(32.0/3.0)
    c2 = 2.0 * (4.0**(1.0/alpha))
    c3 = 6.0 * tilde_C_alpha(alpha)
    term1 = c1*math.sqrt(2.0 * xdet)
    term2 = c2*nu_lux * (math.log((c3*n)/delta)**(1.0/alpha))
    return term1 + term2

def beta_ks_bandit(alpha: float, sigma: float, K: float, U: float,
                   V: np.ndarray, lam: float, delta: float, n: int) -> float:
    """
    K&S(Ber) baseline with the *same* det-term as above for fairness, then add A*x with A=(3/4)K z.
    We enforce x ≥ x_lb_logs(n, \tilde C_α) as in your Prop. analysis.
    """
    A = A_KS(alpha, U=U, K=K, sigma=sigma)
    C_tilde = tilde_C_alpha(alpha)
    sign, logdetV = np.linalg.slogdet(V)
    if sign <= 0:
        raise RuntimeError("Design matrix V must be SPD.")
    d = V.shape[0]
    logdetLamI = d*math.log(lam)
    xdet = 0.5*(logdetV - logdetLamI) - math.log(delta)
    xdet = max(xdet, 1e-12)
    x_conf = max(xdet, x_lb_logs(n, C_tilde))
    c1 = math.sqrt(32.0/3.0)
    return c1*math.sqrt(2.0*xdet) + A * x_conf

# ==========================
# == Bandit main routine  ==
# ==========================
# [Paper linkage] Implements two OFUL-style policies differing only in the confidence radius:
#   • Freedman–ψ_α: β^{(Weib)} from Eq. (beta-oful-weibull)
#   • K&S(Ber):    σ√(2x) + A x with A=(3/4)Kz, same determinant term, same failure level
#   Both share the same {X_t, η_t} path → paired comparison as in your tables.

@dataclass
class H2HResult:
    regrets_F: np.ndarray
    regrets_K: np.ndarray
    final_F: float
    final_K: float

def run_two_policies_one_seed(T: int, d: int, alpha: float, nu_lux: float,
                              L: float, S: float, lam: float, delta: float,
                              cands: int, seed_actions: int, seed_noise: int) -> H2HResult:
    """
    OFUL pipeline with two radii:
       - Freedman–ψ_α: β^{(Weib)} (your paper)
       - K&S(Ber) baseline
    Shared {X_t, η_t} across policies → paired comparison.
    """
    rng_theta = np.random.default_rng(seed_actions + 777777)
    theta = rng_theta.normal(size=d); theta /= np.linalg.norm(theta) + 1e-18
    theta *= S

    # Freedman state
    V_F = lam*np.eye(d); b_F = np.zeros(d)
    regret_F = 0.0
    regrets_F = np.zeros(T) if STORE_FULL_TRAJECTORY else None

    # K&S state
    V_K = lam*np.eye(d); b_K = np.zeros(d)
    regret_K = 0.0
    regrets_K = np.zeros(T) if STORE_FULL_TRAJECTORY else None

    for t in range(1, T+1):
        # Shared candidates and noise
        X_t = candidates_for_t(t, d, L, cands, seed_actions)
        eta_t = noise_for_t(t, alpha, nu_lux, seed_noise)

        # Freedman policy
        theta_hat_F = np.linalg.solve(V_F, b_F)
        Z_F = _V_inv_times_X(V_F, X_t)                 # Z_F = V_F^{-1} X_t
        norms_F = np.sqrt(np.sum(X_t * Z_F, axis=1))   # ||x||_{V^{-1}}
        beta_core_F  = beta_weib_freedman(alpha, nu_lux, V_F, lam, delta, t)
        beta_F       = beta_core_F + math.sqrt(lam)*S
        idx_F   = int(np.argmax(X_t @ theta_hat_F + beta_F * norms_F))
        xF      = X_t[idx_F]
        yF      = float(xF @ theta) + eta_t
        V_F    += np.outer(xF, xF);  b_F += xF*yF

        # K&S policy
        theta_hat_K = np.linalg.solve(V_K, b_K)
        Z_K = _V_inv_times_X(V_K, X_t)
        norms_K = np.sqrt(np.sum(X_t * Z_K, axis=1))
        beta_core_KS = beta_ks_bandit(alpha, BANDIT_SIGMA, BANDIT_K, BANDIT_U, V_K, lam, delta, t)
        beta_KS      = beta_core_KS + math.sqrt(lam)*S
        idx_K   = int(np.argmax(X_t @ theta_hat_K + beta_KS * norms_K))
        xK      = X_t[idx_K]
        yK      = float(xK @ theta) + eta_t
        V_K    += np.outer(xK, xK);  b_K += xK*yK

        # oracle regret (paired)
        r_star   = float(X_t[int(np.argmax(X_t @ theta))] @ theta)
        regret_F += (r_star - float(xF @ theta))
        regret_K += (r_star - float(xK @ theta))
        if STORE_FULL_TRAJECTORY:
            regrets_F[t-1] = regret_F
            regrets_K[t-1] = regret_K

        if (t % 5000) == 0:
            print(f"[t={t}] cumReg(Freedman)={regret_F:.2f} | cumReg(K&S)={regret_K:.2f}")

    if not STORE_FULL_TRAJECTORY:
        regrets_F = np.array([regret_F])
        regrets_K = np.array([regret_K])

    return H2HResult(regrets_F, regrets_K, regret_F, regret_K)

def run_bandit_suite():
    print("\n=== Running paired bandit suite (Freedman–ψ_α vs K&S(Ber)) ===")
    seeds = np.arange(N_SEEDS)
    finals_F, finals_K = [], []
    traj_F, traj_K = [], []

    for s in seeds:
        res = run_two_policies_one_seed(
            T=T_ROUNDS, d=D_BANDIT, alpha=ALPHA, nu_lux=NU,
            L=L_BOUND, S=S_NORM, lam=LAMBDA, delta=(DELTA or 1e-6),
            cands=CANDS_PER_ROUND,
            seed_actions=123456 + 1000*s,
            seed_noise=987654 + 2000*s
        )
        finals_F.append(res.final_F)
        finals_K.append(res.final_K)
        if STORE_FULL_TRAJECTORY:
            traj_F.append(res.regrets_F)
            traj_K.append(res.regrets_K)

    finals_F = np.array(finals_F); finals_K = np.array(finals_K)
    print(f"\nEnd-of-horizon cumulative regret over {N_SEEDS} seeds:")
    print(f"  Freedman–ψ_α: mean={finals_F.mean():.3f}, std={finals_F.std(ddof=1):.3f}")
    print(f"  K&S(Ber):     mean={finals_K.mean():.3f}, std={finals_K.std(ddof=1):.3f}")
    diffs = finals_K - finals_F
    mean_diff = diffs.mean()
    se_diff   = diffs.std(ddof=1) / (math.sqrt(N_SEEDS) if N_SEEDS > 1 else 1.0)
    ci_lo, ci_hi = mean_diff - 1.96*se_diff, mean_diff + 1.96*se_diff
    print(f"  Paired diff (K&S − Freedman) = {mean_diff:.2f} (95% CI [{ci_lo:.2f}, {ci_hi:.2f}])")

    if STORE_FULL_TRAJECTORY:
        traj_F = np.vstack(traj_F); traj_K = np.vstack(traj_K)
        meanF = traj_F.mean(axis=0); meanK = traj_K.mean(axis=0)
        plt.figure(figsize=(8.2, 4.6))
        t = np.arange(1, T_ROUNDS+1)
        plt.plot(t, meanF, label=r"Freedman–$\psi_{\alpha}$")
        plt.plot(t, meanK, label="K&S (Ber)")
        plt.xlabel(r"Rounds $t$")
        plt.ylabel("Cumulative pseudo-regret")
        plt.title(rf"Bandits, $\alpha={ALPHA}$, $\nu={NU}$, $d={D_BANDIT}$, $n={T_ROUNDS}$, seeds={N_SEEDS}")
        plt.legend(); plt.grid(alpha=0.3); plt.tight_layout(); plt.show()

# =========================================
# ======== MULTI-RUN PARAMETER LOOP =======
# =========================================
# [Paper linkage] Three test blocks correspond to α = 1.3, 0.9, 2.0 and provide the summaries
# that populate your bandit tables (mean ± s.d., paired differences with 95% CI).

PARAM_SETS = [
    dict(ALPHA=1.3, NU=2.0, V0=40_000.0,  N_TIME=100_000, D_DIM=50, SIGMA_KS=10.0, U_KS=500.0,  K_KS=1500.0),
    dict(ALPHA=0.9, NU=2.0, V0=20_000.0,  N_TIME=100_000, D_DIM=50, SIGMA_KS=10.0, U_KS=8000.0, K_KS=12000.0),
    dict(ALPHA=2.0, NU=2.0, V0=1_000_000.0,N_TIME=100_000, D_DIM=50, SIGMA_KS=10.0, U_KS=200.0,  K_KS=1000.0),
]

if RUN_BANDITS:
    for i, p in enumerate(PARAM_SETS, 1):
        ALPHA = p["ALPHA"]; NU = p["NU"]; V0 = p["V0"]; N_TIME = p["N_TIME"]; D_DIM = p["D_DIM"]
        SIGMA_KS = p["SIGMA_KS"]; U_KS = p["U_KS"]; K_KS = p["K_KS"]
        BANDIT_SIGMA = SIGMA_KS; BANDIT_U = U_KS; BANDIT_K = K_KS
        print(f"\n=== TEST {i} === (alpha={ALPHA}, nu={NU}, V0={V0}, N_TIME={N_TIME}, D_DIM={D_DIM}, "
              f"sigma={SIGMA_KS}, U={U_KS}, K={K_KS})")
        run_bandit_suite()
```

# <u> **Covariance Simulation**</u>

```python
# [Paper linkage]
# • Section "High-dimensional covariance under heavy tails" (§ simulation_2)
# • Data model: coordinates ∼ sign·Exp(1)^{1/α}, rescaled to unit variance → Σ = I_d
# • Trimming: w_i = min{1, τ / ||X_i - \bar X||}, τ = q-quantile of centered norms
# • Plug-in predictable variation proxy:  V̂ = T − (S S)/n, then v̂ = λ_max( (V̂+V̂^T)/2 )
# • Increment cap for Freedman plug-in: L := τ^2 + ||Σ̂_trim||
# • Fixed-time matrix Freedman tail (diagnostic): d·exp{ − n^2 r^2 / [8 ( v̂ + (L n r)/6 )] } ≤ δ/2
#   (matches the “Freedman piece” structure in your Theorem, with heuristic previsible proxies)
# • K&S(Ber) radius (diagnostic): uses β:=α/2 after squaring, K = L, U ≈ √n·K, σ = √v̂, same failure
#   level; linear term includes z = (4 log(eU/σ))^{1/β}, consistent with your comparison section.

!nvidia-smi
%pip install -U cupy-cuda12x

try:
    import cupy as cp
    if cp.cuda.runtime.getDeviceCount() == 0:
        raise RuntimeError("No CUDA device")
except Exception:
    import numpy as cp

import sys, time, math
import statistics as _stats
import numpy as np
import matplotlib.pyplot as plt
from dataclasses import dataclass

print("Array backend:", "CuPy" if getattr(cp, "__name__", "") == "cupy" else "NumPy")

def opnorm_from_sym(M):
    """
    Operator norm for symmetric/Hermitian M (||M|| = max |λ_i|).
    We use this for ||Σ̂ - Σ|| and also inside the increment cap L.
    """
    evals = cp.linalg.eigvalsh(M)
    return float(cp.max(cp.abs(evals)).item())

def wilson_ci(k, n, z=1.96):
    """Wilson interval for a proportion (coverage reporting)."""
    if n == 0:
        return (0.0, 0.0)
    phat = k / n
    denom = 1.0 + (z*z)/n
    center = (phat + (z*z)/(2*n)) / denom
    half = z * np.sqrt((phat*(1-phat))/n + (z*z)/(4*n*n)) / denom
    return (max(0.0, center - half), min(1.0, center + half))

def sample_chunk(m, d, alpha, b, rng):
    """
    ψ_α sampling in R^d: coordinates are sign * Exp(1)^{1/α}, scaled so Var=1 via b.
    This matches the ‘heavy-tailed but finite variance’ synthetic model.
    """
    u = rng.random((m, d), dtype=cp.float64)
    s = cp.where(u < 0.5, -1.0, 1.0)
    e = rng.standard_exponential(size=(m, d), dtype=cp.float64)
    X = b * s * cp.power(e, 1.0/alpha)
    return X

def run_one_rep(n, d, alpha, qclip, delta, rep_seed, mem_threshold_bytes=5_500_000_000):
    """
    One Monte-Carlo replicate (PLUG-IN comparison):
      1) Generate X ∈ R^{n×d} with ψ_α coordinates (variance normalized).
      2) Trim by radial weight w_i = min{1, τ / ||X_i - \bar X||}, with τ the q-quantile of centered norms.
      3) Compute Σ̂_raw, Σ̂_trim, and a *plug-in* proxy Vhat for the “variance” scale.
      4) Solve the fixed-time matrix Freedman tail   d·exp{- n^2 r^2 / [8( v̂ + (L n r)/6 ) ] } ≤ δ/2
         for r by bisection.      **This is heuristic** (v̂, L are data-derived and not previsible).
      5) Build the K&S(Ber) radius using β:=α/2 after squaring, with K=L and U≈√n K, at the same failure level.
      6) Add a small trimming bias correction and record empirical coverage.
    """
    rng_master = cp.random.default_rng(rep_seed)
    # variance normalizer for Exp(1)^{1/α}: use math.gamma (not np.gamma)
    b_coord = (math.gamma(1.0 + 2.0/alpha))**(-0.5)
    target_Sigma = cp.eye(d, dtype=cp.float64)

    # Chunking to avoid OOM on GPU/CPU
    bytes_needed = n * d * cp.dtype(cp.float64).itemsize
    use_chunks = bytes_needed > mem_threshold_bytes

    if use_chunks:
        # --- First pass: mean ---
        mu = cp.zeros((d,), dtype=cp.float64)
        remaining, idx, CH = n, 0, min(100_000, n)
        while remaining > 0:
            m = min(CH, remaining)
            rng = cp.random.default_rng(rep_seed + 10_000_000 * idx + 7)
            Xc = sample_chunk(m, d, alpha, b_coord, rng)
            mu += cp.sum(Xc, axis=0)
            remaining -= m; idx += 1
        mu /= n

        # --- Second pass: norms for trimming threshold τ ---
        U = cp.empty((n,), dtype=cp.float64)
        remaining, idx, pos = n, 0, 0
        while remaining > 0:
            m = min(CH, remaining)
            rng = cp.random.default_rng(rep_seed + 10_000_000 * idx + 7)
            Xc = sample_chunk(m, d, alpha, b_coord, rng) - mu
            U[pos:pos+m] = cp.linalg.norm(Xc, axis=1)
            remaining -= m; idx += 1; pos += m
        # τ = qclip-quantile of centered norms
        tau = float(cp.quantile(U, qclip).item())
        # weights and a crude bias estimate (we keep the same as in your previous notebook)
        w = cp.minimum(1.0, tau / (U + 1e-12))
        bias_hat = float(cp.mean((1.0 - w*w) * (U*U)).item())

        # --- Third pass: accumulate raw/trimmed scatter and Vhat proxy ---
        S_raw = cp.zeros((d, d), dtype=cp.float64)
        S     = cp.zeros((d, d), dtype=cp.float64)
        T     = cp.zeros((d, d), dtype=cp.float64)
        remaining, idx, pos = n, 0, 0
        while remaining > 0:
            m = min(CH, remaining)
            rng = cp.random.default_rng(rep_seed + 10_000_000 * idx + 7)
            Xc = sample_chunk(m, d, alpha, b_coord, rng) - mu
            Uc = U[pos:pos+m]
            wc = cp.minimum(1.0, tau / (Uc + 1e-12))
            Z  = Xc * wc[:, None]           # trimmed vectors
            S_raw += Xc.T @ Xc              # ∑ X_i X_i^T
            S     += Z.T  @ Z               # ∑ (w_i X_i)(w_i X_i)^T
            norms2 = cp.sum(Z*Z, axis=1)
            T     += Z.T @ (norms2[:, None] * Z)  # ad-hoc proxy towards second moments
            remaining -= m; idx += 1; pos += m

        Shat_raw  = S_raw / n
        Shat_trim = S / n
        # Vhat is a *plug-in* proxy; it is not a literal predictable variation
        Vhat = T - (S @ S) / n

    else:
        rng = rng_master
        X = sample_chunk(n, d, alpha, b_coord, rng)
        X = X - cp.mean(X, axis=0, keepdims=True)
        U = cp.linalg.norm(X, axis=1)
        tau = float(cp.quantile(U, qclip).item())
        w = cp.minimum(1.0, tau / (U + 1e-12))
        Z = X * w[:, None]
        Shat_raw  = (X.T @ X) / n
        S         = Z.T @ Z
        Shat_trim = S / n
        norms2    = cp.sum(Z*Z, axis=1)
        T         = Z.T @ (norms2[:, None] * Z)
        Vhat      = T - (S @ S) / n
        bias_hat  = float(cp.mean((1.0 - w*w) * (U*U)).item())

    # Operator-norm errors
    evals_raw  = cp.linalg.eigvalsh(Shat_raw - target_Sigma)
    evals_trim = cp.linalg.eigvalsh(Shat_trim - target_Sigma)
    op_err_raw  = float(cp.max(cp.abs(evals_raw)).item())
    op_err_trim = float(cp.max(cp.abs(evals_trim)).item())

    # Symmetrize the proxy and extract a scalar scale v̂
    Vhat  = 0.5 * (Vhat + Vhat.T)
    v_hat = float(cp.linalg.eigvalsh(Vhat).max().item())
    # Increment cap (plug-in): L := τ^2 + ||Σ̂_trim||
    L = (tau ** 2) + opnorm_from_sym(Shat_trim)

    # Fixed-time matrix Freedman tail (heuristic plug-in):
    #   P{λ_max(S_n) ≥ n r} ≤ d · exp{ − n^2 r^2 / [8 ( v̂ + (L n r)/6 )] }.
    # We keep the 1/6 (conservative) factor from your proof sketch.
    d_dim = d
    target = delta / 2.0

    def freedman_rhs(r):
        num = (n**2) * (r**2)
        den = 8.0 * (v_hat + (L * n * r) / 6.0)
        return d_dim * np.exp(-num / max(den, 1e-18))

    # Bisection for r_F (expand hi until RHS ≤ target)
    lo, hi = 0.0, 1.0
    while freedman_rhs(hi) > target and hi < 1e6:
        hi *= 2.0
    for _ in range(60):
        mid = 0.5 * (lo + hi)
        if freedman_rhs(mid) <= target:
            hi = mid
        else:
            lo = mid
    rF = hi
    rF_total = rF + bias_hat  # add trimming bias correction

    # K&S(Ber) radius (β := α/2 after squaring), at the *same* failure level:
    beta = max(1e-12, alpha/2.0)
    K_   = L
    Upar = np.sqrt(n) * K_
    log_arg = max(1e-12, np.e * Upar / max(np.sqrt(max(v_hat, 1e-18)), 1e-18))
    z_par  = (4.0 * np.log(log_arg)) ** (1.0 / beta)
    x_par  = max(1.0, np.log((d_dim * n) / max(delta/2.0, 1e-300)))
    s_bound = np.sqrt(max(v_hat, 0.0)) * np.sqrt(2.0 * x_par) + 0.75 * K_ * z_par * x_par
    rK = s_bound / n
    rK_total = rK + bias_hat

    covered_F = 1 if op_err_trim <= rF_total else 0
    covered_K = 1 if op_err_trim <= rK_total else 0

    return dict(op_err_raw=op_err_raw, op_err_trim=op_err_trim,
                rF_total=rF_total, rK_total=rK_total,
                covered_F=covered_F, covered_K=covered_K)

def run_scenario(name, alpha, d, n_grid, delta, qclip, reps, base_seed=12345):
    """
    Scenario runner: reports mean errors, average radii, coverage rates (Wilson CIs), and runtime.
    **Reminder**: The “Freedman r” here is a *plug-in* diagnostic (non-previsible inputs),
    intended for empirical comparison rather than a certified application of the theorem.
    """
    print(f"\n=== Scenario: {name} | alpha={alpha}, d={d}, delta={delta}, qclip={qclip}, reps={reps} ===")
    for n in n_grid:
        errs_raw, errs_trim, rF_list, rK_list = [], [], [], []
        covF = 0; covK = 0
        t0 = time.time()
        for r in range(reps):
            rep_seed = base_seed + 1_000_000_000 * (hash(name) & 0xFFFF) + 10_000 * n + r
            out = run_one_rep(n=n, d=d, alpha=alpha, qclip=qclip, delta=delta, rep_seed=int(rep_seed))
            errs_raw.append(out["op_err_raw"]); errs_trim.append(out["op_err_trim"])
            rF_list.append(out["rF_total"]);    rK_list.append(out["rK_total"])
            covF += out["covered_F"];           covK += out["covered_K"]

        def mean_sem(x):
            m = _stats.fmean(x)
            s = (cp.asarray(x).std().item() / np.sqrt(len(x))) if len(x) > 1 else 0.0
            return m, s

        m_raw,  se_raw  = mean_sem(errs_raw)
        m_trim, se_trim = mean_sem(errs_trim)
        m_rF,   _       = mean_sem(rF_list)
        m_rK,   _       = mean_sem(rK_list)
        covF_rate = covF / reps
        covK_rate = covK / reps
        covF_lo, covF_hi = wilson_ci(covF, reps)
        covK_lo, covK_hi = wilson_ci(covK, reps)

        t1 = time.time()
        print(
            f"n={n:9d} | mean||Σ̂-Σ|| raw={m_raw:.3f}±{se_raw:.3f} | "
            f"trim={m_trim:.3f}±{se_trim:.3f} | "
            f"Freedman r (plug-in)={m_rF:.3f} [cov {covF_rate:.3f}, CI {covF_lo:.3f}-{covF_hi:.3f}] | "
            f"K&S r={m_rK:.3f} [cov {covK_rate:.3f}, CI {covK_lo:.3f}-{covK_hi:.3f}] "
            f"| time {t1-t0:.1f}s")

scenarios = [
    dict(name="(1) alpha=2.0, d=150, q=0.99",
         alpha=2.0, d=150,
         n_grid=[300, 750, 1500, 7500, 30_000, 1_000_000],
         delta=0.05, qclip=0.99, reps=1000),

    dict(name="(2) alpha=1.0, d=150, q=0.95",
         alpha=1.0, d=150,
         n_grid=[300, 750, 1500, 7500, 30_000, 1_000_000],
         delta=0.05, qclip=0.95, reps=2000),

    dict(name="(3) alpha=0.5, d=400, q=0.90",
         alpha=0.5, d=400,
         n_grid=[800, 2000, 4000, 20_000, 80_000, 250_000],
         delta=0.05, qclip=0.90, reps=2000),
]

for sc in scenarios:
    run_scenario(**sc)
```
